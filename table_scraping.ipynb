```python
def url_to_df(url) -> pd.DataFrame:
    """
    引数で渡されたurlの表からデータフレームを作成する
    """
    df = pd.read_html(url)[0]

    # 偶数行 / 奇数行を分ける(コードと市場区分で1社につき2行生成されてしまう)
    df_level_0 = df[df.index % 2 == 1]
    df_level_1 = df[df.index % 2 == 0]

    # multi index columnsをなおす
    df_level_0.columns = df.columns.droplevel(0)
    df_level_1.columns = df.columns.droplevel(1)

    # マージする
    df_merged = pd.concat([df_level_0.reset_index(drop=True), df_level_1.reset_index(drop=True)], axis=1)

    # 重複したカラムを削除する
    df_merged = df_merged.loc[:, ~df_merged.columns.duplicated()].reset_index(drop=True)
    return df_merged
```

```python
def main():

    url_year_dict = {
        "2022": "https://www.jpx.co.jp/listing/stocks/new/index.html",
        "2021": "https://www.jpx.co.jp/listing/stocks/new/00-archives-01.html",
        "2020": "https://www.jpx.co.jp/listing/stocks/new/00-archives-02.html",
        "2019": "https://www.jpx.co.jp/listing/stocks/new/00-archives-03.html",
        "2018": "https://www.jpx.co.jp/listing/stocks/new/00-archives-04.html",
    }

    for year, url in url_year_dict.items():
        df = url_to_df(url)
        df.to_csv(f"{year}.csv", index=False)


if __name__ == "__main__":
        main()
```
