# まとめ

```python
# まとめたコード
## テーブルの抽出
dfs = tabula.read_pdf("（株）坪田ラボ.pdf", lattice=True, pages = '158-159')
## 空データフレーム
new_df = pd.DataFrame(index=[], columns=[])
## データフレームの結合
for df in dfs:
    new_df = pd.concat([df, new_df], axis=0)
new_df
```

```python
## カラム名変更
lcols = list(new_df)

newCols = []

for lcol in lcols:
    new_lcol = lcol.replace('\r','')
    newCols.append(new_lcol)
    
new_df.columns = newCols
df2 = new_df
df2
```

```python
## 所有株式数の処理
df2 = pd.concat([df2, df2["所有株式数(株)"].str.split('\r', expand=True)], axis=1).drop('所有株式数(株)', axis=1)
df2.rename(columns={0: '所有株式数(株)', 1: '削除列'}, inplace=True)
df2.drop("削除列", axis = 1, inplace=True)
```

```python
## 株式の総数に対する所有株式数の割合(%)の処理
df2 = pd.concat([df2, df2["株式(自己株式を除く。)総数に対する所有株式数の割合(%)"].str.split('\r', expand=True)], axis=1).drop('株式(自己株式を除く。)総数に対する所有株式数の割合(%)', axis=1)
df2.rename(columns={0: '株式(自己株式を除く。)総数に対する所有株式数の割合(%)', 1: '削除列'}, inplace=True)
df2.drop("削除列", axis = 1, inplace=True)
```

```python
## 氏名の処理
df2 = pd.concat([df2, df2["氏名又は名称"].str.split('\r', expand=True)], axis=1).drop('氏名又は名称', axis=1)
df2.rename(columns={0: '氏名又は名称', 1: '削除列'}, inplace=True)
df2.drop("削除列", axis = 1, inplace=True)
```

```python
## 列の並び替え
df2 = df2[['氏名又は名称', '住所', '所有株式数(株)' , '株式(自己株式を除く。)総数に対する所有株式数の割合(%)']]
## 余計な行の削除
df2 = df2[~(df2['氏名又は名称'] == '計')]
df2 = df2[~(df2['氏名又は名称'] == '合計')]
df2['株式(自己株式を除く。)総数に対する所有株式数の割合(%)'] = df2['株式(自己株式を除く。)総数に対する所有株式数の割合(%)'].astype(float)
df2 = df2.sort_values('株式(自己株式を除く。)総数に対する所有株式数の割合(%)', ascending=False)
df2
```

```python
# excel出力
df2.to_excel('/Users/mac-user/Desktop/SMTB/業務関連/stock_holders/坪田ラボ.xlsx', index = None)
```
