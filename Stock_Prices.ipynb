```python
# 証券コードとのマッチング用
company = pd.read_excel('/Users/mac-user/Desktop/SMTB/業務関連/newlisted_company/listed_company.xlsx')
```

```python
company['code'][1]
```

```python
# 株価を取得する関数
def get_stock_prices():
    for i in company['code']:
        try:
            ticker_symbol= str(i)
            ticker_symbol_dr=ticker_symbol + ".T"
            ## 取得を開始する日付
            start='2013-01-01'
            end = dt.date.today()
            #データ取得
            df = web.DataReader(ticker_symbol_dr, data_source='yahoo', start=start,end=end)
            #2列目に銘柄コード追加
            df.insert(0, "code", ticker_symbol, allow_duplicates=False)
            #csv保存
            df.to_csv('stock_data_' + ticker_symbol + '.csv')
            print(ticker_symbol + " done")
            time.sleep(1)
        except Exception as e:
            print(ticker_symbol + " error")
            print(e)
            time.sleep(1)
```
